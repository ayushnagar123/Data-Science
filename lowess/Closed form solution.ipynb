## Closed form solution in linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import make_regression
```

### generate dataset

```python
X,Y = make_regression(n_samples=400,n_features=1,n_informative=1,noise=1.8,random_state=11)

Y=Y.reshape((-1,1))# vextor of n*1
```

### normalisation

```python
X = (X-X.mean())/X.std()
```

### Vizualize

```python
plt.style.use("seaborn")
plt.scatter(X,Y)
plt.title("normalized data")
plt.show()
```

```python
ones = np.ones((400,1))
X_ = np.hstack((X,ones))# X shall be m*2 to take product of X*theta
```

```python
def predict(X,theta):
    return np.dot(X,theta)
def getThetaUsingClosedForm(X,Y):
    Y = np.mat(Y)
    p1 = np.dot(X.T,X)
    p2 = np.dot(X.T,Y)
    theta = np.linalg.pinv(p1)*p2
    return theta
```

```python
theta = getThetaUsingClosedForm(X_,Y)
print(theta)
```

```python
plt.scatter(X,Y,label="dataset")
plt.plot(X,predict(X_,theta),c="red",label="prediction")
plt.legend()
plt.show()
```
