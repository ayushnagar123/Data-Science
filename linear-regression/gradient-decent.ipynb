```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
print(X,Y)
```

# Visualize
Given a Function f(x). we want to find the value of x that minimizes f.

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```

```python
x = 0
lr = 0.1  # step size 
error = []
plt.plot(X,Y)
for i in range(50):# 50 steps
    grad = 2*(x-5)
    x = x-lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
    if(len(error) >2 and abs(error[-1]-error[-2])<10**-7):
        print(x)
        break
    print(x)
plt.show()
    
```
