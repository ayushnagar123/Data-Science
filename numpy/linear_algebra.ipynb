# Random Module

```python
import numpy as np
a = np.arange(10)
print(a)
```

```python
np.random.seed(1)
np.random.shuffle(a)
print(a)
```

```python
a = np.random.rand(2,3)
print(a)

a = np.random.randint(5,10,3)
print(a)
```

# statiscs using numpy

## min

```python
a = np.array([[1,2,3,4],[7,6,2,0]])
print(a)
print(np.min(a))
print(np.min(a,axis=0))
print(np.min(a,axis=1))
```

```python
b = np.array([1,2,3,4,5])
m = sum(b)/5
print(m)

print(np.mean(b))
print(np.mean(a,axis=0))
print(np.mean(a,axis=1))
```

```python
c = np.array([1,5,4,2,0])
print(np.median(c))
```

# mean vs average

```python
print(np.mean(c))

# weight

w=np.array([1,2,3,4,5])
print(np.average(c,weights=w))

w=np.array([1,2,3,4,5])
print(np.average(c,weights=w))

# standard deviation
u = np.mean(c)
mystd = np.sqrt(np.mean(abs(c-u)**2))
print(mystd)

# inbuilt function
dev = np.std(c)
print(dev)

# variance

print(mystd**2)
print(np.var(c))
```

# 7. Norms

* norm(L) of a vectort =sqrt(x1^2 + x2^2+x3^2)

* pth norm (x1^p+x2^p+x**p)^(1/p)

* L2 is called Eucledian norm ||x||

* L1 norm is absolute sum of all members of a vector

* Max norm L infinity simplifies absolute value of the element with highest magnitude.

```python
x = np.array([-5,3,10])
lp2 = np.linalg.norm(x)
print(lp2)
```

```python
lp1 = np.linalg.norm(x,ord=1)
print(lp1)
```

```python
lpinf = np.linalg.norm(x,ord=np.inf)
print(lpinf)
```
